```python
# filters_demo.ipynb

import cv2
import matplotlib.pyplot as plt
import os
```

```python
# Path to output folder
output_path = "output"

# List of files to show (in order)
filter_outputs = [
    ("Grayscale", "gray.jpg"),
    ("Gaussian Blur", "blur.jpg"),
    ("Canny Edges", "edges.jpg"),
    ("Binary Threshold", "threshold.jpg"),
    ("Laplacian Filter", "laplacian.jpg"),
    ("Histogram Equalization", "histogram_equalized.jpg")
]
```

```python
for title, filename in filter_outputs:
    filepath = os.path.join(output_path, filename)
    img = cv2.imread(filepath, cv2.IMREAD_UNCHANGED)

    if img is None:
        print(f" Could not load {filename}")
        continue

    # Convert to RGB if image is in color (BGR)
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cmap = None
    else:
        cmap = 'gray'

    plt.figure(figsize=(5, 5))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()
```
